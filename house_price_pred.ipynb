# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel("HousePricePrediction.xlsx")
print(data.head())
```

```python
# to know the dimension of the data 'shape' method is used
data.shape
```

# Data preprocessing

```python
# To find the datatypes of the column based on object, int ,float
```

```python
obj=list(data.select_dtypes(include=['object']).columns)
print("Categorical features:",len(obj))

int_=list(data.select_dtypes(include=['int']).columns)
print("Integer features:",len(int_))

float_=list(data.select_dtypes(include=["float"]).columns)
print("Float features:",len(float_))
```

# EDA Process

```python
num_data=data.select_dtypes(include=['number'])

plt.figure(figsize=(12,6))
sns.heatmap(num_data.corr(),
            cmap='viridis',fmt='.3f',
            linewidths=2,annot=True)
```
