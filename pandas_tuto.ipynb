Making backup data_var to prevent data lose during the usage of inplace=True

```python
data_backup=bios.copy()
data_backup
```

```python
 ## bios.drop(4 ,inplace=True)  ## inplace parameter makes the changes permanent(Once Deleted cannot be recovered so create a backup or reload the data)
 bios.drop(columns=['new_charges'])
```
